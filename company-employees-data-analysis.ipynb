# univariate analysis

```python
gen =df['Gender'].value_counts()
gen
```

```python
plt.bar(gen.index ,gen.values ,color ='purple' )
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('count')
```

```python
plt.pie (x=gen ,labels=gen.index ,autopct= "%0.0f%%",explode=[0,0.1] ,shadow=True ,colors=['#7fb3d5','#8e44ad'])
plt.title('Gender ')
plt.legend(loc=1)
```

```python
year=df['Years'].value_counts()
year
```

```python
sns.barplot(x=year.index, y=year.values)
plt.title('Years Destribuation')
plt.xlabel('years')
plt.ylabel('count')
```

```python
plt.pie(x=year ,labels=year.index ,autopct='%0.0f%%' ,shadow=True,explode=[0,0,0,0,0.3,0.2])
plt.title('years distribution')
plt.legend(loc=1)
```

```python
sns.histplot(x=year.index ,weights=year.values,bins=len(year),kde=True)
plt.title('Years Destribuation')
plt.xlabel('years')
plt.ylabel('count')
```
